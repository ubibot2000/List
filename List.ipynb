### 리스트에서의 병합 및 삽입
#### 복수 개 변수의 병합

```python
#p80
x = 12.23
y = 23.34
packing = [x, y]          # packing!
type(packing)
print('Packing:', packing)
[c1, c2] = packing        # unpacking!
print('Unpacking:\nc1:', c1)
print('c2:', c2)
```

#### 복수 개 리스트의 병합

```python
#p81
fruits1 = ['사과', '오렌지', '포도']
fruits2 = ['복숭아', '키위']
allfruits = fruits1 + fruits2
print(allfruits)
```

#### 리스트에 원소 삽입
##### append() 메소드

```python
prime = [2, 3, 5]
prime.append(7)
print(prime)
```

```python
#p82
a = list()            # list()는 빈 리스트 생성, a = []와 동일
for i in range(5):
    a.append(i)
    print(a[i])
```

```python
fruits = ['사과', '귤', '포도']
fruits.append('수박')
print(fruits)
```

##### insert() 메소드

```python
#p83
fruits = ['사과', '오렌지', '포도']
fruits.insert(1, '키위')
print(fruits)
```

##### 리스트의 함축

```python
#p84
mylist = [3, 5, 4, 9, 1, 8, 2, 1]
newlist = [i for i in mylist if (i%2)==0]
print(newlist)
```

### 리스트의 항목 삭제
#### del 명령문으로 삭제

```python
#p85
nums = [0, 1, 2, 3, 4, 5]
del nums[1]
print(nums)
```

```python
nums = [0, 1, 2, 3, 4, 5]
del nums[1:5]
print(nums)
```

#### pop() 함수로 삭제

```python
#p86
nums = [1, 3, 5, 7, 9]
nums.pop()         # 9가 삭제되어 [1, 3, 5, 7]로 변경됨
nums.pop()         # 7이 삭제되어 [1, 3, 5]로 변경됨
print(nums)        # 결과를 출력
```

```python
nums = [1, 3, 5, 7, 9]
nums.pop(2)
print(nums)
```

#### 빈 리스트로 삭제

```python
nums = [0, 1, 2, 3, 4, 5]
nums[1:2] = []
print(nums)
```
